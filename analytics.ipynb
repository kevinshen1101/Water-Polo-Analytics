```python
import numpy as np
import pandas as pd
```

```python
eger_barceloneta = pd.read_excel("game_data/Eger-Barceloneta.xlsx")
jug_szolnoki = pd.read_excel("game_data/Jug-Szolnok.xlsx")
olympiacos_spandau = pd.read_excel("game_data/Olympiacos-Spandau.xlsx")
prorecco_anbrescia = pd.read_excel("game_data/Pro_Recco- AN_Brescia.xlsx")
spandau04_anbrescia = pd.read_excel("game_data/Spandau_04-AN_Brescia.xlsx")
eger_szolnoki = pd.read_excel("game_data/ZF_Eger-Szolnoki.xlsx")

frames = [eger_barceloneta, 
          jug_szolnoki,
          olympiacos_spandau,
          prorecco_anbrescia,
          spandau04_anbrescia,
          eger_szolnoki]

df = pd.concat(frames, join="outer")
print(df)
```

```python
'''
Shooter analytics
'''

scoring_plays = df.loc[df['Goal'] == 1]

print("Shooting Player Goal Counts")
goal_counts = scoring_plays.groupby('Shooter Position').count()['Goal']
print(goal_counts)

print("Shooting Player sot Counts")
shot_counts = df.groupby('Shooter Position').count()['Goal']
print(shot_counts)

print("SHOOTING EFFICIENCY")
print(goal_counts.divide(shot_counts))
```

```python
'''
Assists Analytics
'''

scoring_plays = df.loc[df['Goal'] == 1]

print("Assiting Player Goal Counts")
shot_counts = df.groupby('Assisting Player Position').count()['Goal']
print(shot_counts)

print("Assiting Player Goal Counts")
goal_counts = scoring_plays.groupby('Assisting Player Position').count()['Goal']
print(goal_counts)

print("Assisting Player Goal Percentages")
print(goal_counts.divide(shot_counts, fill_value=0))
```

```python
'''
Shooter/Assister pair
'''

total_pairs = df.copy()
total_pairs ['Shooter/Assister pair'] = list(zip(total_pairs['Shooter Position'], total_pairs['Assisting Player Position']))
scoring_plays = df.loc[df['Goal'] == 1]
scoring_plays['Shooter/Assister pair'] = list(zip(scoring_plays['Shooter Position'], scoring_plays['Assisting Player Position']))

print("Shooter/Assister Pair Goal Counts")
print(scoring_plays.groupby('Shooter/Assister pair').count()['Goal'])
```

```python
'''
Number of passes analytics
'''

scoring_plays = df.loc[df['Goal'] == 1]

print("Number of Passes Shot Counts")
shot_counts = df.groupby('Number of Passes').count()['Goal']
print(shot_counts)

print("Number of Passes Goal Counts")
goal_counts = scoring_plays.groupby('Number of Passes').count()['Goal']
print(goal_counts)

print("Number of Passes Percentages")
print(goal_counts.divide(shot_counts, fill_value=0))
```

```python
'''
Misc analytics
'''

scoring_plays = df.loc[df['Goal'] == 1]
print("Overall 6-on-5 Goal Percentage: "+str(scoring_plays.shape[0]/df.shape[0]))

dropped_plays = df.loc[df['Dropped Ball'] == 1]
scoring_plays = dropped_plays.loc[dropped_plays['Goal'] == 1]
print("Overall 6-on-5 Goal Percentage: "+str(scoring_plays.shape[0]/dropped_plays.shape[0]))

timeout_plays = df.loc[df['Timeout'] == 1]
scoring_plays = timeout_plays.loc[timeout_plays['Goal'] == 1]
print("Overall Timeout Goal Percentage: "+str(scoring_plays.shape[0]/timeout_plays.shape[0]))

backcourt_plays = df.loc[df['Kickout Position'] == 1]
scoring_plays = backcourt_plays.loc[backcourt_plays['Goal'] == 1]
print("Overall Backcourt Kickout Goal Percentage: "+str(scoring_plays.shape[0]/backcourt_plays.shape[0]))

frontcourt_plays = df.loc[df['Kickout Position'] == 0]
scoring_plays = frontcourt_plays.loc[frontcourt_plays['Goal'] == 1]
print("Overall Frontcourt Kickout Goal Percentage: "+str(scoring_plays.shape[0]/frontcourt_plays.shape[0]))
```
